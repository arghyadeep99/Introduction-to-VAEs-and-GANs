```
import torch
import torch.nn
import torch.nn.functional as nn
import torch.autograd as autograd
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
from torch.autograd import Variable
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([i/100 for i in range(100000)])
y = 2*x + 5 + np.array([np.random.rand()*300 for i in range(100000)])
XX = np.array([x,y]).T
XX = XX/np.max(XX)
plt.scatter(XX.T[0],XX.T[1])
XX = torch.Tensor(XX)
```

```
mb_size = 100
z_dim = 2
X_dim = 2
h_dim = 100
cnt = 0
lr = 1e-4
```

```
G = torch.nn.Sequential(
    torch.nn.Linear(z_dim, h_dim),
    torch.nn.ReLU(),
    torch.nn.Linear(h_dim, X_dim),
    torch.nn.Sigmoid()
)
```

```
D = torch.nn.Sequential(
    torch.nn.Linear(X_dim, h_dim),
    torch.nn.ReLU(),
    torch.nn.Linear(h_dim, 1),
)
```

```
def reset_grad():
    G.zero_grad()
    D.zero_grad()
```

```
G_solver = optim.RMSprop(G.parameters(), lr=lr)
D_solver = optim.RMSprop(D.parameters(), lr=lr)
```

```
G_losses = []
D_losses = []
for it in range(5000):
  
    # Generator forward-loss-backward-update
    z = Variable(torch.randn(mb_size, z_dim))
    for _ in range(5):
        # Sample data
        perm = torch.randperm(100000)
        idx = perm[:mb_size]
        samples = XX[idx]
        X = samples

        # Dicriminator forward-loss-backward-update
        G_sample = G(z)
        D_real = D(X)
        D_fake = D(G_sample)

        D_loss = -(torch.mean(D_real) - torch.mean(D_fake))

        D_loss.backward()
        D_solver.step()

        # Weight clipping
        for p in D.parameters():
            p.data.clamp_(-0.01, 0.01)

        # Housekeeping - reset gradient
        reset_grad()

    G_sample = G(z)
    D_fake = D(G_sample)

    G_loss = -torch.mean(D_fake)

    G_loss.backward()
    G_solver.step()

    # Housekeeping - reset gradient
    reset_grad()

    # Print and plot every now and then
    if it % 200 == 0:
        print('Iter-{}; D_loss: {}; G_loss: {}'
              .format(it, D_loss.data.numpy(), G_loss.data.numpy()))

        samples = G(z).data.numpy()[:16]

        fig = plt.figure(figsize=(4, 4))
        gs = gridspec.GridSpec(4, 4)
        gs.update(wspace=0.05, hspace=0.05)

        print('G_Loss: ',G_loss.detach().numpy(),'D_Loss: ',D_loss.detach().numpy())
        G_losses.append(G_loss.detach().numpy())
        D_losses.append(D_loss.detach().numpy())
        plt.scatter(X.numpy().T[0],X.numpy().T[1])
        plt.scatter(G_sample.detach().numpy().T[0],G_sample.detach().numpy().T[1])
        plt.show()

        # visual discriminator
        mat = []
        for x in range(100):
          mat.append([])
          for y in range(100):
            mat[-1].append([x/100,y/100])
        p = D(torch.tensor(mat,dtype=torch.float32))
        p = p.detach().numpy().reshape(100,100)
        plt.imshow(p)
        plt.colorbar()
        plt.show()
        
            
```

```
import torch
import torch.nn
import torch.nn.functional as nn
import torch.autograd as autograd
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
from random import random

x1 = 0.2
y1 = 0.2
x2 = 0.5
y2 = 0.5
x3 = 0.8
y3 = 0.8

x = np.array([x1+random()/3 for i in range(10000)]+[x2+random()/3 for i in range(10000)]+[x3+random()/3 for i in range(10000)])
y = np.array([y1+random()/4 for i in range(10000)]+[y2+random()/3 for i in range(10000)]+[y3+random()/4 for i in range(10000)])
XX = np.array([x,y]).T
XX = XX/np.max(XX)
XX -= np.min(XX)
np.random.shuffle(XX)
plt.scatter(XX.T[0],XX.T[1])
XX = torch.Tensor(XX)


mb_size = 1000
z_dim = 10
X_dim = 2
h_dim = 100
cnt = 0
lr = 1e-4

G = torch.nn.Sequential(
    torch.nn.Linear(z_dim, h_dim),
    torch.nn.ReLU(),
    torch.nn.Linear(h_dim, X_dim),
    torch.nn.Sigmoid()
)

D = torch.nn.Sequential(
    torch.nn.Linear(X_dim, h_dim),
    torch.nn.ReLU(),
    torch.nn.Linear(h_dim, 1),
)

def reset_grad():
    G.zero_grad()
    D.zero_grad()
    
G_solver = optim.RMSprop(G.parameters(), lr=lr)
D_solver = optim.RMSprop(D.parameters(), lr=lr)
```

```
G_losses = []
D_losses = []
for it in range(10000):
  
    # Generator forward-loss-backward-update
    z = Variable(torch.randn(mb_size, z_dim))
    for _ in range(5):
        # Sample data
        samples = XX[it:it+100000]
        X = samples

        # Dicriminator forward-loss-backward-update
        G_sample = G(z)
        D_real = D(X)
        D_fake = D(G_sample)

        D_loss = -(torch.mean(D_real) - torch.mean(D_fake))

        D_loss.backward()
        D_solver.step()

        # Weight clipping
        for p in D.parameters():
            p.data.clamp_(-0.01, 0.01)

        # Housekeeping - reset gradient
        reset_grad()

    G_sample = G(z)
    D_fake = D(G_sample)

    G_loss = -torch.mean(D_fake)

    G_loss.backward()
    G_solver.step()

    # Housekeeping - reset gradient
    reset_grad()

    # Print and plot every now and then
    if it % 200 == 0:
        print('Iter-{}; D_loss: {}; G_loss: {}'
              .format(it, D_loss.data.numpy(), G_loss.data.numpy()))

        samples = G(z).data.numpy()[:16]

        fig = plt.figure(figsize=(4, 4))
        gs = gridspec.GridSpec(4, 4)
        gs.update(wspace=0.05, hspace=0.05)

        print('G_Loss: ',G_loss.detach().numpy(),'D_Loss: ',D_loss.detach().numpy())
        G_losses.append(G_loss.detach().numpy())
        D_losses.append(D_loss.detach().numpy())
        plt.scatter(X.numpy().T[0],X.numpy().T[1])
        plt.scatter(G_sample.detach().numpy().T[0],G_sample.detach().numpy().T[1])
        plt.show()

        # visual discriminator
        mat = []
        for x in range(100):
          mat.append([])
          for y in range(100):
            mat[-1].append([x/100,(100-x)/100])
        p = D(torch.tensor(mat,dtype=torch.float32))
        p = p.detach().numpy().reshape(100,100)
        plt.imshow(p)
        plt.colorbar()
        plt.show()
        
            
```
